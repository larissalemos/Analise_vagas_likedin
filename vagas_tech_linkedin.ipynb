```python
# Importanto bibliotecas

import pandas as pd
import numpy as np
```

```python
# Importando a base de dados

df = pd.read_csv('scrapped data.csv')
df.head(3)
```

```python
# Separando coluna Level_and_involvement em duas colunas

df[['Level', 'Involvement']] = df["Level_and_involvement"].str.split("·", expand=True)
df = df.drop(columns='Level_and_involvement')
df.head(3)
```

```python
# Separando coluna Industry_and_Employee_count em duas colunas

df[['Employees', 'Industry']] = df["Industry_and_Employee_count"].str.split("·", expand=True)
df = df.drop(columns='Industry_and_Employee_count')
df.head(3)
```

```python
# Separando coluna Location em três colunas

df[['Cidade', 'Estado', 'País']] = df["Location"].str.split(",", expand=True)
df = df.drop(columns='Location')
df.head(3)
```

```python
# Retirando palavras desnecessárias das colunas de seguidores, candidatos, descrição da vaga

df['LinkedIn_Followers'] = df['LinkedIn_Followers'].str.replace('followers', '')
df['LinkedIn_Followers'] = df['LinkedIn_Followers'].str.replace('follower', '')
df['LinkedIn_Followers'] = df['LinkedIn_Followers'].str.replace(',', '')
df['Total_applicants'] = df['Total_applicants'].str.replace('applicants', '')
df['Total_applicants'] = df['Total_applicants'].str.replace('applicant', '')
df['Employees'] = df['Employees'].str.replace('employees', 'colaboradores')
df['job_description'] = df['job_description'].str.replace('Role Purpose \n', '')

df.head(3)
```

```python
# Identificando valores únicos da coluna Level
valores_unicos_level = df['Level'].unique()
valores_unicos_level
```

```python
# Traduzindo valores da coluna Level
df['Level'] = df['Level'].str.replace('Full-time', 'Tempo integral')
df['Level'] = df['Level'].str.replace('Part-time', 'Meio período')
df['Level'] = df['Level'].str.replace('Contract', 'Contrato')
df
```

```python
# Identificando valores únicos da coluna Involvement
valores_unicos_involvement = df['Involvement'].unique()
valores_unicos_involvement
```

```python
# Traduzindo valores da coluna Involvement
df['Involvement'] = df['Involvement'].str.replace('Entry level', 'Junior')
df['Involvement'] = df['Involvement'].str.replace('Mid-Senior level', 'Pleno')
df['Involvement'] = df['Involvement'].str.replace('Director', 'Diretoria')
df['Involvement'] = df['Involvement'].str.replace('Associate', 'Associado')
df.head()
```

```python
# Identificando valores únicos da coluna Industry
valores_unicos_industry = df['Industry'].unique()
valores_unicos_industry
```

```python
# Traduzindo valores da coluna Industry
df['Industry'] = df['Industry'].str.replace('IT Services and IT Consulting', 'Serviços e consultoria em TI')
df['Industry'] = df['Industry'].str.replace('Technology, Information and Internet', 'Tecnologia da informação')
df['Industry'] = df['Industry'].str.replace('Real Estate', 'Imobiliário')
df['Industry'] = df['Industry'].str.replace('Telecommunications', 'Telecomunicações')
df.head()
```

```python
df.info()
```

```python
# Preenchendo linhas vazias com "0" e transformando em tipo inteiro

df['Total_applicants'] = df['Total_applicants'].fillna(0)
df['Total_applicants'] = df['Total_applicants'].astype('int')
df['LinkedIn_Followers'] = df['LinkedIn_Followers'].fillna(0)
df['LinkedIn_Followers'] = df['LinkedIn_Followers'].astype('int')
df.info()
```

```python
# Calculando a média do número de seguidores no LinkedIn agrupando por empresa

data = df[['Name', 'LinkedIn_Followers']].groupby('Name').mean().reset_index()

dict_data = dict(zip(data['Name'], data['LinkedIn_Followers']))

for i in dict_data:
    df.loc[df['Name'] == i, 'LinkedIn_Followers'] = int(dict_data[i])

df.head()
```

```python
# Checando número de valores nulos em cada coluna

nulos = df.isnull().sum()

print(nulos)
```

```python
# Removendo linhas com valores nulos

df.dropna(inplace=True)
df.reset_index(inplace=True, drop=True)
df.shape
```

```python
df.columns
```

```python
# Traduzindo o nome das colunas para português

df.columns = ['titulo', 'empresa', 'descricao', 'candidatos',
       'seguidores_linkedin', 'dedicacao', 'senioridade', 'numero_colaboradores', 'setor',
       'cidade', 'estado', 'pais']

df.head()
```

### Identificando habilidades técnicas mencionadas na descrição da vaga

```python
# Listando as habilidades técnicas que queremos encontrar
habilidades = ['PYTHON', 'PANDAS', 'NUMPY', 'PYTORCH', 
                    'POWER BI', 'TABLEAU', 'EXCEL',
                    'SQL', 'MYSQL', 'POSTGRESQL', 'MONGODB', 'AI', 'ML']
```

```python
# Criando colunas para as habilidades listadas e atribuindo o valor 0 a todas as linhas

df[habilidades] = 0
df.head(3)
```

```python
# Extraindo habilidades da descrição da vaga

habilidades_desc = df['descricao'].str.upper().apply(lambda x: [i.upper() for i in habilidades if i in x])
habilidades_desc
```

```python
# Atribuindo o valor 1 nas colunas de habilidades que foram mencionadas na descrição

for index, habilidades_lista in enumerate(habilidades_desc):
    for x in habilidades_lista:
        df[x][index] = 1

df.head(5)
```

```python
# Encontrando colunas com todas as linhas zeradas e eliminá-las

zeros = (df == 0) & (df.applymap(type) == int)
cols = zeros.all()[zeros.all()].index.to_list()
df = df.drop(cols, axis=1)

df.shape
```

```python
# Excluindo a coluna de descrição

df = df = df.drop(columns = 'descricao', axis=1)

df.head()
```

```python
df['titulo'] = df['titulo'].str.upper()
df
```

```python
valores_unicos_titulo = df['titulo'].unique()
valores_unicos_titulo
```

```python
df['titulo'] = df['titulo'].str.replace('OPENING FOR ', '')
df['titulo'] = df['titulo'].str.replace('LTIMINDTREE', '')
df['titulo'] = df['titulo'].str.replace('-', ' ')
df['titulo'] = df['titulo'].str.replace('.', '')
df['titulo'] = df['titulo'].str.replace(',', '')
df['titulo'] = df['titulo'].str.replace('/', ' ')
df['titulo'] = df['titulo'].str.replace('SENIOR', '')
df['titulo'] = df['titulo'].str.replace('JUNIOR', '')
df['titulo'] = df['titulo'].str.replace('LOOKING FOR', '')
df['titulo'] = df['titulo'].str.replace('INDIA', '')
df['titulo'] = df['titulo'].str.replace('(3-5 YEARS)', '')
df['titulo'] = df['titulo'].str.replace('BHUBANESHWAR', '')
df['titulo'] = df['titulo'].str.replace('IN LTIM', '')
df['titulo'] = df['titulo'].str.replace('JR.', '')
df['titulo'] = df['titulo'].str.replace('REMOTE', '')
df['titulo'] = df['titulo'].str.replace('(3 5 YEARS)', '')
df['titulo'] = df['titulo'].str.replace('(INSURETECH COMPANY)', '')
df['titulo'] = df['titulo'].str.replace('(WITH LEADING BANK)', '')
df['titulo'] = df['titulo'].str.replace('(5 8 YRS) INDORE', '')
df['titulo'] = df['titulo'].str.replace('PAID', '')
df['titulo'] = df['titulo'].str.replace('HIRING', '')
df['titulo'] = df['titulo'].str.replace('PAN', '')
df['titulo'] = df['titulo'].str.replace('SAP', '')
df['titulo'] = df['titulo'].str.replace("'", '')
df['titulo'] = df['titulo'].str.strip()
```
